```python
# convert date columns to dt
df_green_2019['lpep_pickup_datetime'] = pd.to_datetime(df_green_2019['lpep_pickup_datetime'])
df_green_2019['lpep_dropoff_datetime'] = pd.to_datetime(df_green_2019['lpep_dropoff_datetime'])
df_green_2020['lpep_pickup_datetime'] = pd.to_datetime(df_green_2020['lpep_pickup_datetime'])
df_green_2020['lpep_dropoff_datetime'] = pd.to_datetime(df_green_2020['lpep_dropoff_datetime'])
df_green_2021['lpep_pickup_datetime'] = pd.to_datetime(df_green_2021['lpep_pickup_datetime'])
df_green_2021['lpep_dropoff_datetime'] = pd.to_datetime(df_green_2021['lpep_dropoff_datetime'])
```

```python
green_2019_mean = df_green_2019['fare_amount'].mean()
green_2019_std = df_green_2019['fare_amount'].std()
green_2019_right = green_2019_mean + (green_2019_std * 3)
green_2019_left = green_2019_mean - (green_2019_std * 3)

df_green_2019.drop(df_green_2019[df_green_2019['fare_amount'] < 1].index, inplace = True)
df_green_2019.drop(df_green_2019[df_green_2019['fare_amount'] > green_2019_right].index, inplace = True)

df_green_2019 = df_green_2019.sort_values('fare_amount', ascending=False)

green_2020_mean = df_green_2020['fare_amount'].mean()
green_2020_std = df_green_2020['fare_amount'].std()
green_2020_right = green_2020_mean + (green_2020_std * 3)
green_2020_left = green_2020_mean - (green_2020_std * 3)

df_green_2020.drop(df_green_2020[df_green_2020['fare_amount'] < 1].index, inplace = True)
df_green_2020.drop(df_green_2020[df_green_2020['fare_amount'] > green_2020_right].index, inplace = True)

df_green_2020 = df_green_2020.sort_values('fare_amount', ascending=False)

green_2021_mean = df_green_2021['fare_amount'].mean()
green_2021_std = df_green_2021['fare_amount'].std()
green_2021_right = green_2021_mean + (green_2021_std * 3)
green_2021_left = green_2021_mean - (green_2021_std * 3)

df_green_2021.drop(df_green_2021[df_green_2021['fare_amount'] < 1].index, inplace = True)
df_green_2021.drop(df_green_2021[df_green_2021['fare_amount'] > green_2021_right].index, inplace = True)

df_green_2021 = df_green_2021.sort_values('fare_amount', ascending=False)
```

1. What has been the economic impact of the pandemic on NYC taxi revenue over time?

```python
green_rev_2019 = df_green_2019['total_amount'].sum() - df_green_2019['tip_amount'].sum()
green_rev_2020 = df_green_2020['total_amount'].sum() - df_green_2020['tip_amount'].sum()
green_rev_2021 = df_green_2021['total_amount'].sum() - df_green_2021['tip_amount'].sum()

print('Green Taxi Revenue')
print('2019 Total Fare Revenue: $  {: 9,.2f}'.format(green_rev_2019))
print('2020 Total Fare Revenue: $  {: 9,.2f}'.format(green_rev_2020))
print('2021 Total Fare Revenue: $  {: 9,.2f}'.format(green_rev_2021))
```

Calculate the percent change from 2019 to 2020

```python
green_pct_chg = (green_rev_2020 - green_rev_2019) / green_rev_2019
print('{:.2%}'.format(green_pct_chg))
```
