```python
yellow_2020_mean = df_yellow_2020['fare_amount'].mean()
yellow_2020_std = df_yellow_2020['fare_amount'].std()
yellow_2020_right = yellow_2020_mean + (yellow_2020_std * 3)
yellow_2020_left = yellow_2020_mean - (yellow_2020_std * 3)

df_yellow_2020.drop(df_yellow_2020[df_yellow_2020['fare_amount'] < 1].index, inplace = True)
df_yellow_2020.drop(df_yellow_2020[df_yellow_2020['fare_amount'] > yellow_2020_right].index, inplace = True)

df_yellow_2020 = df_yellow_2020.sort_values('fare_amount', ascending=False)


yellow_2021_mean = df_yellow_2021['fare_amount'].mean()
yellow_2021_std = df_yellow_2021['fare_amount'].std()
yellow_2021_right = yellow_2021_mean + (yellow_2021_std * 3)
yellow_2021_left = yellow_2021_mean - (yellow_2021_std * 3)

df_yellow_2021.drop(df_yellow_2021[df_yellow_2021['fare_amount'] < 1].index, inplace = True)
df_yellow_2021.drop(df_yellow_2021[df_yellow_2021['fare_amount'] > yellow_2021_right].index, inplace = True)

df_yellow_2021 = df_yellow_2021.sort_values('fare_amount', ascending=False)
```

## Analysis

1. What has been the economic impact of the pandemic on NYC taxi revenue over time?

Not including taxes, tips, and surcharges

```python
yellow_rev_2019 = df_yellow_2019['total_amount'].sum() - df_yellow_2019['tip_amount'].sum()
yellow_rev_2020 = df_yellow_2020['total_amount'].sum() - df_yellow_2020['tip_amount'].sum()
yellow_rev_2021 = df_yellow_2021['total_amount'].sum() - df_yellow_2021['tip_amount'].sum()

print('Yellow Taxi Revenue')
print('2019 Total Fare Revenue: ${: 9,.2f}'.format(yellow_rev_2019))
print('2020 Total Fare Revenue: $  {: 9,.2f}'.format(yellow_rev_2020))
print('2021 Total Fare Revenue: $  {: 9,.2f}'.format(yellow_rev_2021))
```

Calculate the percent change from 2019 to 2020

```python
yellow_pct_chg = (yellow_rev_2020 - yellow_rev_2019) / yellow_rev_2019
print('{:.2%}'.format(yellow_pct_chg))
```

3. What's the most popular payment method? Did this change because of the pandemic?

```python
# find most common payment methods in 2020
print(df_yellow_2019['payment_type'].mode())
print(df_yellow_2020['payment_type'].mode())
print(df_yellow_2021['payment_type'].mode())
```

```python
# count len payment_type without the NaN values
total_pay_yellow_2019 = (np.count_nonzero(~np.isnan(df_yellow_2019['payment_type'])))
total_pay_yellow_2020 = (np.count_nonzero(~np.isnan(df_yellow_2020['payment_type'])))
```

```python
# Yellow payment types
print('Percent of Fares Paid W/ Card 2019: {:.2%}'.format(len(df_yellow_2019.loc[df_yellow_2019['payment_type'] == 1]) / total_pay_yellow_2019))
print('Percent of Fares Paid W/ Card 2020: {:.2%}'.format(len(df_yellow_2020.loc[df_yellow_2020['payment_type'] == 1]) / total_pay_yellow_2020))

print('Percent of Fares Paid W/ Cash 2019: {:.2%}'.format(len(df_yellow_2019.loc[df_yellow_2019['payment_type'] == 2]) / total_pay_yellow_2019))
print('Percent of Fares Paid W/ Cash 2020: {:.2%}'.format(len(df_yellow_2020.loc[df_yellow_2020['payment_type'] == 2]) / total_pay_yellow_2020))
```
